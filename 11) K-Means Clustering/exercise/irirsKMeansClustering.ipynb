```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
dir(iris)
```

```python
iris.data
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data)
```

```python
df.head()
```

```python
df.drop(df[[0,1]],axis='columns', inplace=True)
```

```python
df
```

```python
plt.scatter(df[2],df[3])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[[2]])
df[2] = scaler.transform(df[[2]])
```

```python
scaler.fit(df[[3]])
df[3] = scaler.transform(df[[3]])
```

```python
plt.scatter(df[2],df[3])
```

```python
df
```

```python
from sklearn.cluster import KMeans


def analysis(n):
   sse =[]
   for i in range(n):
       km = KMeans(n_clusters=(i+1))
       km.fit(df)
       sse.append(km.inertia_)
   return sse
    


    
```

```python
sse_res = analysis(11)
print(sse_res)
```

```python
sse_res
```

```python
plt.plot(sse_res)
```
