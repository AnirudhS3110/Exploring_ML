```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_digits
dig = load_digits()
```

```python
dir(dig)
```

```python
df = pd.DataFrame(dig.data, columns=dig.feature_names)
df
```

```python
df['target'] = dig.target
```

```python
df.head(10)
```

```python
import numpy as np
```

```python
df.shape
```

Normalising the values!

```python
x = df.drop('target',axis='columns')
```

```python
y = df.target
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

## Now lets just train the model normally without dividing eh dataset into principal Components

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
model1.fit(x_train,y_train)
model1.score(x_test,y_test)
```

Now lets use Principal COmponent Analysis. 

```python
from sklearn.decomposition import PCA
pc = PCA(.95)
```

```python
x_new = pc.fit_transform(x_scaled)
```

```python
x_new.shape
```

x had 69 rows, now it has only 40 rows

Now lets again split the ds and train the model
|


```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_new,y, test_size=0.2, random_state=30)
```

```python
model2 = LogisticRegression()
model2.fit(x_train,y_train)
model2.score(x_test,y_test)
```

## now lets have only 2 Principal Components

```python
pc2 = PCA(n_components=2)
x_new2 = pc2.fit_transform(x_scaled)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_new2,y, test_size=0.2, random_state=30)
```

```python
model3 = LogisticRegression()
model3.fit(x_train,y_train)
model3.score(x_test,y_test)
```

### As we have only 2 Principal Components, we are getting only 56% accuracy
