```python
import pandas as pd
df = pd.read_csv("heights.csv")
df.head()
```

```python
right_threshold = df['height'].quantile(.95)
right_threshold
```

```python
left_threshold = df['height'].quantile(0.05)
left_threshold
```

### Now we are removing the outliers from extreme right and left ends. This technique requires domain knowledge to implement this method.

```python
required_df = df[(df['height'] > left_threshold) & (right_threshold > df['height'])]
required_df 
```
