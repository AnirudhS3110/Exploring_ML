```python
import pandas as pd
df = pd.read_csv("bhp.csv")
df.head(10)
```

```python
min_threshold, max_threshold = df['price_per_sqft'].quantile([0.001 , 0.999])
min_threshold , max_threshold
```

```python
df[df.price_per_sqft < min_threshold]
```

### When you analyse the above DataFrame, you can see 9 BHK for 416 rupees per SQFt, which is truly an outlier

```python
df[df.price_per_sqft > max_threshold]
```

```python
cleaned_df = df[(df.price_per_sqft> min_threshold) & (df.price_per_sqft < max_threshold)]
cleaned_df.head()
```
