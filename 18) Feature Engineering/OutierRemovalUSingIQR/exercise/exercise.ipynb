```python
import pandas as pd
df = pd.read_csv("height_weight.csv")
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import norm
```

```python
plt.hist(df.height, bins=40, rwidth=0.7,density=True)
plt.xlabel("height")
plt.ylabel("count")
rng = np.arange(df.height.min(), df.height.max(), 0.1)
plt.plot(rng, norm.pdf(rng, df.height.mean(), df.height.std()))
plt.show()
```

```python
plt.hist(df.weight, bins=40, rwidth=0.7, density=True)
plt.xlabel("weight")
plt.ylabel("count")
rng = np.arange(df.weight.min(), df.weight.max(), 0.1)
plt.plot(rng, norm.pdf(rng, df.weight.mean(), df.weight.std()))
plt.show()
```

## The Weight Curve Doesnt matches the Bell Curve..

Now Detecting IQR for Weight and Height

```python
def findMinMax(obj):
    Q1 = obj.quantile(0.25)
    Q3 = obj.quantile(0.75)
    IQR = Q3-Q1
    return Q1 - 1.5*IQR , Q3+1.5*IQR

```

```python
min_height, max_height = findMinMax(df.height)
```

```python
min_weight,max_weight = findMinMax(df.weight)
```

```python
print("Outliers of Weight:")
df[(df.weight < min_weight) | (df.weight > max_weight)]
```

```python
print("Outliers of Weight:")
df[(df.height < min_height) | (df.height > max_height)]
```
