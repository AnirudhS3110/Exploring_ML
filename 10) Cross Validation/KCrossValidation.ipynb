```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
digits = load_digits()
x_train,x_test,y_train,y_test = train_test_split(digits.data,digits.target, train_size=0.8)
```

```python
model1 = LinearRegression()
model1.fit(x_train,y_train)
model1.score(x_test,y_test)
```

```python
model2 = RandomForestClassifier()
model2.fit(x_train,y_train)
model2.score(x_test,y_test)
```

```python
model3 = DecisionTreeClassifier()
model3.fit(x_train,y_train)
model3.score(x_test,y_test)
```

```python
model4 = RandomForestClassifier(n_estimators = 130)
model4.fit(x_train,y_train)
model4.score(x_test,y_test)
```

```python
model5 = SVC()
model5.fit(x_train,y_train)
model5.score(x_test,y_test)
```

- The Disadvantage of using train_test_split is everytime we train the model, the Model gets trained by different permutations of Dataset
- So we are gonna try use KCrossValidation

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits = 3)
kf
```

```python
for train_index, test_index in kf.split([1 ,2, 3, 4, 5, 6, 7, 8, 9]):
    print(train_index,test_index)
```

```python
def get_score(model,x_train,x_test,y_train,y_test):
    model.fit(x_train,y_train)
    return model.score(x_test,y_test)
    
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
sk = StratifiedKFold(n_splits=3)
```

```python
logsitic_score = []
svc_score = []
randomforest_score = []
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
for train_index,test_index in sk.split(digits.data,digits.target):
    x_train,x_test,y_train,y_test = digits.data[train_index],digits.data[test_index],digits.target[train_index],digits.target[test_index]
    logsitic_score.append(get_score(LogisticRegression(max_iter=10000),x_train,x_test,y_train,y_test))
    svc_score.append(get_score(SVC(),x_train,x_test,y_train,y_test))
    randomforest_score.append(get_score(RandomForestClassifier(),x_train,x_test,y_train,y_test))
```

## The whole sk.split() for loop can be replaced by inbuilt function called cross_val_score

```python
from sklearn.model_selection import cross_val_score
l_score = cross_val_score(LogisticRegression(max_iter=1000),digits.data,digits.target)
svc_score= cross_val_score(SVC(),digits.data,digits.target )
randomforest_score = cross_val_score(RandomForestClassifier(),digits.data,digits.target)
```
