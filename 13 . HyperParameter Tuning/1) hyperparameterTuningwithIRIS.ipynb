```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df
```

```python
df['target'] = iris.target
```

```python
df.head(10)
```

```python
x = df.drop('target', axis='columns')
```

```python
x.head(40)
```

```python
y = df.target
```

Creating the SVC model:

```python
from sklearn.svm import SVC
model = SVC(kernel = 'rbf', C=30,gamma='auto')
```

#### Lets Revise Cross Validation Code:

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits=5)

for train_index,test_index in kf.split(x,y):
    #print(train_index,test_index)
    scores = []
    x_train,x_test,y_train,y_test = x.iloc[train_index],x.iloc[test_index], y.iloc[train_index],y.iloc[test_index]  #iloc is ot locatet the rows in the dataframe
    model.fit(x_train,y_train)
    print(model.score(x_test,y_test))
    
    
```

### Now Lets do with cross_val_score,the inbuilt function which directly gives the Result:

```python
from sklearn.model_selection import cross_val_score
cross_val_score(SVC(kernel='rbf', gamma ='auto', C=10), x,y, cv=5)
```

```python
cross_val_score(SVC(kernel='rbf', gamma ='auto', C=30), x,y,cv=5)
```

```python
cross_val_score(SVC(kernel='rbf', gamma ='auto', C=40), x,y,cv=5)
```

For the above method, we can find the avg score by taking avewrage

#### The above method seems time consuming as its more manual and repetitive, there is sklearn library called GridSearchCV to do it!

```python
from sklearn.model_selection import GridSearchCV
```

Syntax: GridSearchCV( model , parameters that yuou wanna cutomize in json like sructure.)

```python
clf = GridSearchCV(SVC(gamma='auto'), {
    'kernel':['rbf','linear','poly'],
    'C':[10,30,40]
}, cv=5, return_train_score=False) #cv implies how many Cross validation needs to run
```

```python
clf.fit(x,y)
clf.cv_results_
```

```python
CrossValidation_result = pd.DataFrame(clf.cv_results_)
CrossValidation_result
```

### We dont need to view all the values in the table, we need to view which values of C and Kernel gives optimal result, so:

```python
CrossValidation_result[['param_C','param_kernel','mean_test_score']]
```

```python
dir(CrossValidation_result)
```

From the Table we can notice with C= 10 and Kernel = RBF, we are getting the best Result among others

### One of the Problem which happens in GridSearchCV is computation Cost
### To tackle this COmputation Problem sklearn Library comes up with: **RandomizedSearchCV**

```python
from sklearn.model_selection import RandomizedSearchCV
rs= RandomizedSearchCV(SVC(gamma='auto'),{
    'kernel': ['rbf','linear'],
    'C':[1,10,30]
},
cv=5, return_train_score=False, n_iter=3)
rs.fit(x,y)
cv_res = pd.DataFrame(rs.cv_results_)[['param_C','param_kernel','mean_test_score']]
cv_res
```
