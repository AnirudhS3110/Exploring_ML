```python
import pandas as pd
```

```python
df = pd.read_csv('spam.csv')
```

```python
df.head()
```

```python
df['spam'] = df['Category'].apply( lambda x: 1 if x == 'ham' else 0 )
```

```python
df.head()
```

```python
df.drop('Category',axis='columns',inplace =True)
```

```python
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.Message, df.spam, train_size= .75)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
x_train_count =v.fit_transform(x_train)
#x_train_count.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
model.fit(x_train_count,y_train)
```

```python
def transformm(x):
    input = v.transform(x)
    return input
```

```python
model.score(transformm(x_test), y_test)
```

```python
model.predict(transformm(x_test[:5]))
```

```python
chatgpt_test = ["Hey, can we meet tomorrow to discuss the project?",
"Earn $1000 a day working from home, no experience needed!",
"I’ll call you in the evening to confirm the details.",
"You’ve been selected for a luxury vacation giveaway!",
"Exclusive deal: Buy 1 get 3 free, only today!"]
```

```python
model.predict(transformm(chatgpt_test))
```

### We dont need to write codes for transforming a text to tabular form of numbers, there are pipelines in sklearn where we can define a pipeline of Transformation.

```python
from sklearn.pipeline import Pipeline
```

```python
model1 = Pipeline([
('Vectorize', CountVectorizer()),
('nb', MultinomialNB())])
```

```python
model1.fit(x_train,y_train)
```

```python
model1.predict(chatgpt_test)
```
